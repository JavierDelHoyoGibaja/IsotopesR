```python
import pandas as pd
import numpy as np
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Cargar el archivo All.xlsx
file_path = r"C:\Users\jdelhoyo\PhD\Study cases\Genissiat\Arve-Valserine\Isotopes\GitHub Isotopes\Data\OnlyDataIsotopes.xlsx"
df = pd.read_excel(file_path)

# Mostrar información básica del archivo
print("Dimensiones del archivo:", df.shape)
print("\nPrimeras filas:")
print(df.head(10))
print("\nNombres de las columnas:")
print(df.columns.tolist())
print("\nTipos de datos:")
print(df.dtypes)
```

```python
# Explorar categorías únicas en Sp y Cond
print("Categorías únicas en 'Sp':")
print(df['Sp'].unique())
print(f"Conteo: {df['Sp'].value_counts()}")

print("\nCategorías únicas en 'Cond':")
print(df['Cond'].unique())
print(f"Conteo:\n{df['Cond'].value_counts()}")

print("\nCrosstab de Sp y Cond:")
print(pd.crosstab(df['Sp'], df['Cond']))
```

```python

# ==========================================
# CREACIÓN DE 6 GRUPOS PRINCIPALES
# ==========================================

# FRAXINUS: 3 grupos
frax_living_float = df[(df['Sp'] == 'Fraxinus') & (df['Cond'].isin(['Living', 'Float']))]
frax_living_out = df[(df['Sp'] == 'Fraxinus') & (df['Cond'].isin(['Living', 'Out']))]
frax_living_sub = df[(df['Sp'] == 'Fraxinus') & (df['Cond'].isin(['Living', 'Sub']))]

# ALNUS: 3 grupos
alnus_living_float = df[(df['Sp'] == 'Alnus') & (df['Cond'].isin(['Living', 'Float']))]
alnus_living_out = df[(df['Sp'] == 'Alnus') & (df['Cond'].isin(['Living', 'Out']))]
alnus_living_sub = df[(df['Sp'] == 'Alnus') & (df['Cond'].isin(['Living', 'Sub']))]

# Crear un diccionario con todos los grupos para fácil acceso
grupos = {
    'Fraxinus_Living_Float': frax_living_float,
    'Fraxinus_Living_Out': frax_living_out,
    'Fraxinus_Living_Sub': frax_living_sub,
    'Alnus_Living_Float': alnus_living_float,
    'Alnus_Living_Out': alnus_living_out,
    'Alnus_Living_Sub': alnus_living_sub
}

# Mostrar estadísticas de cada grupo
print("=" * 70)
print("ESTADÍSTICAS DE LOS 6 GRUPOS")
print("=" * 70)

for nombre_grupo, grupo_df in grupos.items():
    print(f"\n{nombre_grupo}:")
    print(f"  - Número de muestras: {len(grupo_df)}")
    print(f"  - Condiciones incluidas: {grupo_df['Cond'].unique().tolist()}")
    print(f"  - D (δD): mean={grupo_df['D'].mean():.2f}, min={grupo_df['D'].min():.2f}, max={grupo_df['D'].max():.2f}")
    print(f"  - O (δO): mean={grupo_df['O'].mean():.2f}, min={grupo_df['O'].min():.2f}, max={grupo_df['O'].max():.2f}")
```

```python

# ==========================================
# MANN-KENDALL TEST (CORRECTED)
# ==========================================
# For each group: ONE single test on the complete time series
# Step 1: Calculate mean value per Order (chronological time point)
# Step 2: Apply Mann-Kendall test to the complete ordered time series
# Result: One tau and p-value per group for D, one per group for O

from scipy.stats import kendalltau

print("=" * 100)
print("MANN-KENDALL TEST: ONE TEST PER GROUP")
print("Single monotonic trend test on complete time series (group-level analysis)")
print("=" * 100)

# Store results for summary
mk_results_group = {}

for nombre_grupo, grupo_df in grupos.items():
    print(f"\n{'='*100}")
    print(f"GRUPO: {nombre_grupo}")
    print(f"{'='*100}")
    
    # Step 1: Group by Order and calculate mean D and O for each time point
    time_series = grupo_df.groupby('Order')[['D', 'O']].mean().reset_index()
    time_series = time_series.sort_values('Order')
    
    # Extract time indices and values
    time_points = time_series['Order'].values
    mean_d = time_series['D'].values
    mean_o = time_series['O'].values
    
    n_timepoints = len(time_series)
    
    print(f"\nTime series summary:")
    print(f"  Number of time points (Order values): {n_timepoints}")
    print(f"  Time points: {time_points.tolist()}")
    print(f"\n{'Order':<8} {'Mean D (δD)':<18} {'Mean O (δO)':<18}")
    print("-" * 50)
    for i, order_val in enumerate(time_points):
        print(f"{int(order_val):<8} {mean_d[i]:<18.6f} {mean_o[i]:<18.6f}")
    
    # Step 2: Apply Mann-Kendall test to the complete time series
    # Using time indices (0, 1, 2, ...) as the time variable
    time_indices = np.arange(n_timepoints)
    
    tau_d, p_d = kendalltau(time_indices, mean_d)
    tau_o, p_o = kendalltau(time_indices, mean_o)
    
    # Interpret trends
    trend_d = "Increasing" if tau_d > 0 else "Decreasing"
    significant_d = "***" if p_d < 0.05 else "NS"
    
    trend_o = "Increasing" if tau_o > 0 else "Decreasing"
    significant_o = "***" if p_o < 0.05 else "NS"
    
    print(f"\n{'='*50}")
    print("MANN-KENDALL TEST RESULTS")
    print(f"{'='*50}")
    print(f"\nD (δD) - {trend_d}:")
    print(f"  Kendall's tau = {tau_d:.6f}")
    print(f"  p-value (two-sided) = {p_d:.6f}")
    print(f"  Significance: {significant_d}")
    
    print(f"\nO (δO) - {trend_o}:")
    print(f"  Kendall's tau = {tau_o:.6f}")
    print(f"  p-value (two-sided) = {p_o:.6f}")
    print(f"  Significance: {significant_o}")
    
    # Store results
    mk_results_group[nombre_grupo] = {
        'n_timepoints': n_timepoints,
        'D_tau': tau_d,
        'D_p_value': p_d,
        'D_trend': trend_d,
        'O_tau': tau_o,
        'O_p_value': p_o,
        'O_trend': trend_o
    }

# Print comprehensive summary table
print(f"\n\n{'='*100}")
print("SUMMARY TABLE: MANN-KENDALL RESULTS FOR ALL GROUPS")
print(f"{'='*100}")

print(f"\n{'Group':<30} {'Variable':<12} {'N':<6} {'Kendall τ':<12} {'p-value':<12} {'Trend':<15} {'Sig.':<6}")
print("-" * 110)

for nombre_grupo, resultado in mk_results_group.items():
    print(f"{nombre_grupo:<30} {'D':<12} {resultado['n_timepoints']:<6} {resultado['D_tau']:<12.6f} {resultado['D_p_value']:<12.6f} {resultado['D_trend']:<15} {'***' if resultado['D_p_value'] < 0.05 else 'NS':<6}")
    print(f"{'':<30} {'O':<12} {resultado['n_timepoints']:<6} {resultado['O_tau']:<12.6f} {resultado['O_p_value']:<12.6f} {resultado['O_trend']:<15} {'***' if resultado['O_p_value'] < 0.05 else 'NS':<6}")
    print()

print("\n*** indicates p < 0.05 (statistically significant)\nNS indicates not significant (p ≥ 0.05)")
```

```python

# ==========================================
# DESCRIPTIVE STATISTICS: MEAN AND STD BY ORDER
# ==========================================
# For each group and each Order: calculate mean and std dev across samples

print("=" * 120)
print("DESCRIPTIVE STATISTICS BY ORDER: MEAN AND STANDARD DEVIATION")
print("=" * 120)

# Store descriptive statistics
descriptive_stats = {}

for nombre_grupo, grupo_df in grupos.items():
    print(f"\n{'='*120}")
    print(f"GRUPO: {nombre_grupo}")
    print(f"{'='*120}")
    
    # Calculate mean and std by Order for both D and O
    stats_by_order = grupo_df.groupby('Order')[['D', 'O']].agg(['mean', 'std', 'count']).reset_index()
    stats_by_order = stats_by_order.sort_values('Order')
    
    # Flatten column names
    stats_by_order.columns = ['Order', 'D_mean', 'D_std', 'D_count', 'O_mean', 'O_std', 'O_count']
    
    print(f"\n{'Order':<8} {'N_samples':<12} {'D_mean':<15} {'D_std':<15} {'O_mean':<15} {'O_std':<15}")
    print("-" * 90)
    
    for idx, row in stats_by_order.iterrows():
        order_val = int(row['Order'])
        n_samples = int(row['D_count'])
        d_mean = row['D_mean']
        d_std = row['D_std']
        o_mean = row['O_mean']
        o_std = row['O_std']
        
        print(f"{order_val:<8} {n_samples:<12} {d_mean:<15.6f} {d_std:<15.6f} {o_mean:<15.6f} {o_std:<15.6f}")
    
    descriptive_stats[nombre_grupo] = stats_by_order

print(f"\n\n{'='*120}")
print("SUMMARY: CONFIRMATION OF CALCULATIONS")
print(f"{'='*120}")
print("\n✓ CONFIRMED: Mean values have been calculated for each Order")
print("✓ NEW: Standard deviation values calculated for each Order")
print("✓ INTERPRETATION: Std Dev shows the variability between samples at each time point")
print("\nFor example:")
print("  - If all 10 samples at Order 0 have similar D values → Low std dev")
print("  - If the 10 samples at Order 0 have very different D values → High std dev")
```

```python
# ==========================================
# CORRECTED: GROUP-LEVEL EXTREMES ANALYSIS
# ==========================================
# For each of the 6 groups: find max/min by Order (within that group)
# Then calculate deviations from that group's Order 0 baseline

print("=" * 140)
print("GROUP-LEVEL ANALYSIS: MAXIMUM AND MINIMUM VALUES (BY GROUP, NOT BY SPECIES)")
print("=" * 140)

group_extremes_analysis = {}

for nombre_grupo, grupo_df in grupos.items():
    
    print(f"\n{'='*140}")
    print(f"GRUPO: {nombre_grupo}")
    print(f"{'='*140}")
    
    # Step 1: Get time series (mean per Order for THIS group)
    time_series = grupo_df.groupby('Order')[['D', 'O']].mean().reset_index()
    time_series = time_series.sort_values('Order')
    
    mean_d_values = time_series['D'].values
    mean_o_values = time_series['O'].values
    orders = time_series['Order'].values
    
    # Step 2: Get baseline (Order 0 of this specific group)
    baseline_d = time_series[time_series['Order'] == 0]['D'].values[0]
    baseline_o = time_series[time_series['Order'] == 0]['O'].values[0]
    
    print(f"\nBaseline (Order 0 for this group):")
    print(f"  D (δD): {baseline_d:.6f}")
    print(f"  O (δO): {baseline_o:.6f}")
    
    print(f"\nTime series for this group:")
    print(f"{'Order':<8} {'D_mean':<15} {'O_mean':<15}")
    print("-" * 45)
    for i, order_val in enumerate(orders):
        print(f"{int(order_val):<8} {mean_d_values[i]:<15.6f} {mean_o_values[i]:<15.6f}")
    
    # Step 3: Find maximum and minimum
    max_d_idx = np.argmax(mean_d_values)
    min_d_idx = np.argmin(mean_d_values)
    max_o_idx = np.argmax(mean_o_values)
    min_o_idx = np.argmin(mean_o_values)
    
    max_d_value = mean_d_values[max_d_idx]
    max_d_order = int(orders[max_d_idx])
    max_d_deviation = max_d_value - baseline_d
    
    min_d_value = mean_d_values[min_d_idx]
    min_d_order = int(orders[min_d_idx])
    min_d_deviation = min_d_value - baseline_d
    
    max_o_value = mean_o_values[max_o_idx]
    max_o_order = int(orders[max_o_idx])
    max_o_deviation = max_o_value - baseline_o
    
    min_o_value = mean_o_values[min_o_idx]
    min_o_order = int(orders[min_o_idx])
    min_o_deviation = min_o_value - baseline_o
    
    print(f"\n{'='*60}")
    print("EXTREME VALUES FOR THIS GROUP")
    print(f"{'='*60}")
    
    print(f"\nD (δD) - MAXIMUM:")
    print(f"  Order: {max_d_order}")
    print(f"  Value: {max_d_value:.6f}")
    print(f"  Deviation from baseline: {max_d_deviation:+.6f}")
    
    print(f"\nD (δD) - MINIMUM:")
    print(f"  Order: {min_d_order}")
    print(f"  Value: {min_d_value:.6f}")
    print(f"  Deviation from baseline: {min_d_deviation:+.6f}")
    
    print(f"\nO (δO) - MAXIMUM:")
    print(f"  Order: {max_o_order}")
    print(f"  Value: {max_o_value:.6f}")
    print(f"  Deviation from baseline: {max_o_deviation:+.6f}")
    
    print(f"\nO (δO) - MINIMUM:")
    print(f"  Order: {min_o_order}")
    print(f"  Value: {min_o_value:.6f}")
    print(f"  Deviation from baseline: {min_o_deviation:+.6f}")
    
    # Store results
    group_extremes_analysis[nombre_grupo] = {
        'Baseline_D': baseline_d,
        'Baseline_O': baseline_o,
        'Max_D_value': max_d_value,
        'Max_D_order': max_d_order,
        'Max_D_deviation': max_d_deviation,
        'Min_D_value': min_d_value,
        'Min_D_order': min_d_order,
        'Min_D_deviation': min_d_deviation,
        'Max_O_value': max_o_value,
        'Max_O_order': max_o_order,
        'Max_O_deviation': max_o_deviation,
        'Min_O_value': min_o_value,
        'Min_O_order': min_o_order,
        'Min_O_deviation': min_o_deviation
    }

# Print comprehensive summary table
print(f"\n\n{'='*140}")
print("SUMMARY TABLE: GROUP-LEVEL EXTREMES (CORRECTED)")
print(f"{'='*140}")

print(f"\n{'Group':<25} {'Variable':<12} {'Type':<10} {'Order':<8} {'Value':<15} {'Deviation':<15}")
print("-" * 120)

for nombre_grupo, analysis in group_extremes_analysis.items():
    print(f"{nombre_grupo:<25} {'D':<12} {'Maximum':<10} {analysis['Max_D_order']:<8} {analysis['Max_D_value']:<15.6f} {analysis['Max_D_deviation']:+<15.6f}")
    print(f"{'':<25} {'':<12} {'Minimum':<10} {analysis['Min_D_order']:<8} {analysis['Min_D_value']:<15.6f} {analysis['Min_D_deviation']:+<15.6f}")
    print(f"{'':<25} {'O':<12} {'Maximum':<10} {analysis['Max_O_order']:<8} {analysis['Max_O_value']:<15.6f} {analysis['Max_O_deviation']:+<15.6f}")
    print(f"{'':<25} {'':<12} {'Minimum':<10} {analysis['Min_O_order']:<8} {analysis['Min_O_value']:<15.6f} {analysis['Min_O_deviation']:+<15.6f}")
    print()
```
